## Pairwise log-rank tests

```python
#| label: tbl-pairwise
#| tbl-cap: "Pairwise log-rank test results with Bonferroni correction for 3 comparisons."
#| code-fold: true
#| code-summary: "Pairwise log-rank tests"

groups = df['Food'].unique()
bonferroni = 0.05 / 3

pairwise_results = []
for g1, g2 in combinations(groups, 2):
    d1 = df[df['Food'] == g1]
    d2 = df[df['Food'] == g2]
    res = logrank_test(d1['survival_days'], d2['survival_days'], d1['event'], d2['event'])
    pairwise_results.append({
        'Comparison': f'{g1} vs {g2}',
        'Test statistic': round(res.test_statistic, 3),
        'p-value': round(res.p_value, 4),
        'Significant (0.05)': res.p_value < 0.05,
        f'Significant (Bonferroni {bonferroni:.4f})': res.p_value < bonferroni,
    })

pd.DataFrame(pairwise_results)
```
